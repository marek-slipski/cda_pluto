```python
test_sub = hdulist_v2_1['PRIMARY'].data[60:100,710:740]
```

```python
test_sub_g = skfilt.gaussian(test_sub/351.,sigma=0.8)
print test_sub_g.min(),test_sub_g.max()
```

```python
plt.figure(figsize=(6,6))
plt.imshow(test_sub_g,cmap='Greys')
plt.show()
```

```python
test_sub_sob = skfilt.sobel(test_sub/351)
test_sub_g_sob = skfilt.sobel(test_sub_g)
```

```python
fig, ax = plt.subplots(1,2,figsize=(12,6))
ax[0].imshow(test_sub_sob,cmap='Greys')
ax[1].imshow(test_sub_g_sob,cmap='Greys')
plt.show()
```
