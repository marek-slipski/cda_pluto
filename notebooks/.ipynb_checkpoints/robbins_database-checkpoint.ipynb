```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

#####******* don't convert to integers - doesn't have to be pixel, just point in pixel space
```

```python
rdb = pd.read_csv('../Robbins/Supplemental Material - Pluto.csv')
print rdb.columns
```

```python
scale = 69.132491671495  #pixels/deg
colpix = int(np.round(360*scale,0))
rowpix = colpix/2
print colpix, rowpix 
rdb['col'] = rdb['MASTER_LON']/360.*colpix
rdb['row'] = (rdb['MASTER_LAT']/90.*rowpix/2.-rowpix/2)*-1
```

```python
plt.figure()
plt.scatter(rdb['col'],rdb['row'],s=1)
plt.ylim(rowpix,0)
plt.show()
```

```python
plt.figure()
plt.scatter(rdb['col'],rdb['row'],s=1)
plt.ylim(6000,4400)
plt.xlim(8900,10300)
plt.show()
```

```python
test = np.load('../slices/4400_6000_8900_10300.npy')
```

```python
plt.figure(figsize=(10,10))
plt.imshow(test)
plt.plot(rdb['col']-8900,rdb['row']-4400,'ro',ms=5)
plt.xlim(0,1400)
plt.ylim(1600,0)
plt.show()
```

```python
big= rdb[rdb['MASTER_DIAM']>2]
big_range = big[(big['col']>8900)&(big['col']<10300)]
big_range = big[(big['row']>4400)&(big['row']<6000)]
```
