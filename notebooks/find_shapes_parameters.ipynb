```python
one_bound = contours[4].reshape(len(contours[4]),2)
ts_m = cv2.moments(one_bound) #get info on boundary
ts_cx = int(ts_m['m10']/ts_m['m00']) #centroid x
ts_cy = int(ts_m['m01']/ts_m['m00']) #centroid y

print ts_cx, ts_cy
```

```python
dist = np.zeros((len(one_bound))) #init distance
theta = np.zeros((len(one_bound)))
```

```python
for i,(x,y) in enumerate(one_bound):
    xdist = (x-ts_cx)
    ydist = (y-ts_cy)
    if x - ts_cx > 0:
        temp = np.arctan(float((y-ts_cy))/(x-ts_cx))
        if y - ts_cy >= 0:
            theta[i] = temp #between 0 and pi/2
        elif y - ts_cy <0 :
            theta[i] = temp + 2*np.pi #from -pi/2 to 0 --> 3pi/2 to 2pi
    elif x - ts_cx < 0:
        temp = np.arctan(float((y-ts_cy))/(x-ts_cx))
        theta[i] = temp + np.pi #quads 4-1 to 2-3
    else:
        if y - ts_cy > 0:
            theta[i] = np.pi/2
        elif y - ts_cy < 0:
            theta[i] = 3*np.pi/2
    dist[i] = np.sqrt(xdist**2 + ydist**2)
```

```python
def ab_j(j,radial,angle):
    n = len(radial)
    a_j = 2./n * np.sum(radial*np.cos(j*theta))
    b_j = 2./n * np.sum(radial*np.sin(j*theta))
    return a_j, b_j
```

```python
import numpy, scipy, scipy.ndimage, scipy.interpolate, numpy.fft, math

# create simple square
#img = numpy.zeros( (10, 10) )
#img[1:9, 1:9] = 1
#img[2:8, 2:8] = 0
```

```python
C = test_r + test_i*1.j
# find contour
#x, y = numpy.nonzero(img)

# find center point and conver to polar coords
#x0, y0 = numpy.mean(x), numpy.mean(y)
#C = (x - x0) + 1j * (y - y0)
angles = numpy.angle(C)
distances = numpy.absolute(C)
sortidx = numpy.argsort( angles )
angles = angles[ sortidx ]
distances = distances[ sortidx ]

# copy first and last elements with angles wrapped around. needed so can interpolate over full range -pi to pi
angles = numpy.hstack(([ angles[-1] - 2*math.pi ], angles, [ angles[0] + 2*math.pi ]))
distances = numpy.hstack(([distances[-1]], distances, [distances[0]]))

# interpolate to evenly spaced angles
f = scipy.interpolate.interp1d(angles, distances)
angles_uniform = scipy.linspace(-math.pi, math.pi, num=100, endpoint=False) 
distances_uniform = f(angles_uniform)

# fft and inverse fft
fft_coeffs = numpy.fft.rfft(distances_uniform)

# zero out all but lowest 10 coefficients
fft_coeffs[11:] = 0
distances_fit = numpy.fft.irfft(fft_coeffs)

#import matplotlib.pyplot as plt
plt.polar(angles, distances,'b')
plt.polar(angles_uniform, distances_uniform,'g')
plt.polar(angles_uniform, distances_fit,'r')
plt.show()
```

```python
An = np.sqrt(fft_coeffs.real**2+fft_coeffs.imag**2)
plt.figure()
plt.plot(np.arange(len(An)),An)
plt.xlim(1,10)
plt.ylim(0,200)
plt.show()
```
